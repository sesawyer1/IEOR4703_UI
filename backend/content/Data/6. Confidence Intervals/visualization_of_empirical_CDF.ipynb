## Empirical vs True CDF (ECDF vs CDF)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from time import time
```

```python
# import ecdf function
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
np.random.seed(3489872)
```

```python
sample_size = np.array([50, 100, 200, 5000])
size1 = len(sample_size)

plt.figure(figsize=[15,8])

dx = 0.01
lB = -4
uB = 4

for i in range(size1):
    len1 = sample_size[i]
    z = np.random.randn(len1)

    x = np.arange(lB, uB+dx, dx)
    # empirical cdf
    ecdf_z = ECDF(z)
    e_f = ecdf_z(x)

    # actual pdf
    f = sp.stats.norm.cdf(x)

    plt.subplot(1, len(sample_size), i+1)
    plt.plot(x,f,'--')
    plt.plot(x,e_f)

    plt.title('sample size = %i' % len1)
    plt.legend(['ecdf', 'cdf'])

plt.show()
```
