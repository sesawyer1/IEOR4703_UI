### Illustration of early exercise

```python
import numpy as np
from time import time
import matplotlib.pyplot as plt
%matplotlib inline
```

### Parameters

```python
Spot = 100
maturity = 1
sig = 0.15
mu = 0.04
m = 250
dt = maturity/m
```

```python
t = np.linspace(0, maturity, m+1)
z = np.random.randn(m)
delta_log_s = (mu - sig*sig/2)*dt + sig*np.sqrt(dt)*z
log_s = np.zeros(m+1)
log_s[0] = np.log(Spot)
log_s[1:] = delta_log_s
log_s = np.cumsum(log_s)
S = np.exp(log_s)
```

```python
x = np.linspace(-1, 1, 200)
len1 = len(x)
y1 = np.ones_like(x)
y2 = np.sqrt(1 - x**2)
X = np.zeros(2*len1)
X[:len1] = x
X[len1:] = x[::-1]
Y = np.zeros_like(X)
Y[:len1] = y1
Y[len1:] = y2
```

```python
plt.figure()
plt.plot(S, t, color='C0')
plt.axvline(x=Spot, color='C1')
plt.fill(X, Y, color='C2')
plt.show()
```
