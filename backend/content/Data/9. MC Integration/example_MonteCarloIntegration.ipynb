### Estimation of an integral using Monte Carlo simulation

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

We want to estimate $\int_0^{10} -x^5 + 5x^4 + 60x^3 + 7x + 10 \;\mathrm{d}x$.

```python
x = np.arange(0, 10 + 0.1, 0.1)
y = -x**5 + 5*x**4 + 60*x**3 + 7*x + 10
```

```python
plt.figure()
plt.plot(x, y)
plt.show()
```

In this case, we can compute the exact solution $\theta$.

```python
z = 10
theta = (-1/6)*z**6 + z**5 + 15*z**4 + 3.5*z**2 + 10*z
print('theta = ' + str(theta))
```

We approximate the solution via MC.

```python
np.random.seed(6223456)
```

```python
N = 200000
U = np.random.rand(N) # U ~ Uniform(0, 1)
```

If $U\sim\text{Uniform}(0,1)$, then $X = a + (b-a)U$ is $\text{Uniform}(a,b)$.

```python
X = 10*U # U ~ Uniform(0, 10)
Y = -X**5 + 5*X**4 + 60*X**3 + 7*X + 10

theta_Hat = 10 * np.mean(Y)
```

```python
print('theta     = ' + str(theta))
print('theta_Hat = ' + str(theta_Hat))
```
