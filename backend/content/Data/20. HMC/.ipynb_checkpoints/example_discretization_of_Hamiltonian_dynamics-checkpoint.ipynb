# IEOR 4703 -- Monte Carlo Simulation for FE

## An example of the discretization of the Hamiltonian dynamics

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

## Euler Method

```python
n = 500000 # there will be overflow with values larger than 1500

m = 1
e = 0.001

x = np.arange(-1., 1.01, 0.01)
y = np.sqrt(np.abs(1 - x**2))

# Euler Method

q1 = np.zeros(n+1)
p1 = np.zeros(n+1)

q1[0] = 0
p1[0] = 1

for i in range(1, n+1):
    p1[i] = p1[i-1] - e*q1[i-1]
    q1[i] = q1[i-1] + e*p1[i-1]/m

# there can be overflow, so we only keep the finite values
lim = np.max(np.where(np.isfinite(p1)))
q1 = q1[:lim]
p1 = p1[:lim]

plt.figure(figsize=(6, 6))
plt.plot(q1, p1, '.-')
plt.plot(x, y, color='C1', linewidth=2)
plt.plot(x, -y, color='C1', linewidth=2)
plt.xlabel('q')
plt.ylabel('p')
plt.title('Euler Method, $\epsilon$ =' + str(e))
plt.show()
```

## Modified Euler

```python
n = 70

m = 1
e = 0.09

q2 = np.zeros(n+1)
p2 = np.zeros(n+1)

q2[0] = 0
p2[0] = 1

for i in range(1, n+1):
    p2[i] = p2[i-1] - e*q2[i-1]
    q2[i] = q2[i-1] + e*p2[i]/m

plt.figure(figsize=(6, 6))
plt.plot(q2, p2, '.', linewidth=5)
plt.plot(x, y, color='C1', linewidth=2)
plt.plot(x, -y, color='C1', linewidth=2)
plt.xlabel('q')
plt.ylabel('p')
plt.title('Modified Euler Method, $\epsilon$ =' + str(e))
plt.show()
```

## Leapfrog Method

```python
n = 40

m = 1
e = 0.19

q3 = np.zeros(n+1)
p3 = np.zeros(n+1)

q3[0] = 0
p3[0] = 1

for i in range(1, n+1):
    p3[i] = p3[i-1] - 0.5*e*q3[i-1]
    q3[i] = q3[i-1] + e*p3[i]/m
    p3[i] = p3[i] - 0.5*e*q3[i]


plt.figure(figsize=(6, 6))
plt.plot(q3, p3, '.', linewidth=8)
plt.plot(x, y, color='C1', linewidth=2)
plt.plot(x, -y, color='C1', linewidth=2)
plt.xlabel('q')
plt.ylabel('p')
plt.title('Modified Euler Method, $\epsilon$ =' + str(e))
plt.show()
```
