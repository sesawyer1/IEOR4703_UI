# IEOR 4703 -- Monte Carlo Simulation for FE

### simulated paths of stock price evolutions under GBM & time-changed Brownian motion

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S0 = 100
r = 0.05
q = 0.01
sg = 0.2

nu = 0.03
theta = -0.1

T = 1
N = 250
dt = T / N
```

```python
pp1 = np.zeros(N + 1)
pp2 = np.zeros(N + 1)
t = np.arange(0, T + dt, dt)
```

```python
np.random.seed(53672321)
```

```python
fd1 = np.log(S0)
fd2 = np.log(S0)

drf1 = (r - q) * dt + dt/nu*np.log(1.0 - theta *nu - (sg**2)*nu/2)

pp1[0] = S0

z = np.random.randn(N)

for i in range(N):
    fg = np.random.gamma(dt/nu, nu)
    # time-changed Brownian motion
    feed1 = sg * np.sqrt(fg) * z[i]
    fd1 = fd1 + feed1 + drf1
    pp1[i + 1] = np.exp(fd1)
```
