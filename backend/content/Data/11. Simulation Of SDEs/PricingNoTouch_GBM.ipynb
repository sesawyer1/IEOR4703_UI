```python
import numpy as np
import matplotlib.pyplot as plt
import random

# import our BMS price and delta functions
from BMS import BMS_price
```

```python
spot = 1000
maturity = 2
m = 500
N = 20000
dt = maturity/m

sig = 0.30
r = 0.045
q = 0.0125

Ub = 1300
Lb = 700

rebate = 100
K = 1000
```

```python
C = BMS_price('call', spot, K, r, q, sig, maturity)
print('C = ' + str(C))
```

```python
np.random.seed(7331166)
```

```python
runningSum1 = 0
runningSum2 = 0

runningSum1b = 0
runningSum2b = 0

for j in range(N):

    if j%1000 == 0:
        print(j)

    # reseting for each path
    S1 = spot
    S2 = spot

    flag1 = 0
    flag2 = 0


    for i in range(m):

        z = np.random.randn(1)

        #Euler
        S1 = S1 + (r-q)*S1*dt + sig*S1*np.sqrt(dt)*z

        if flag1 == 0:
            if S1 > Ub or S1 < Lb:
                flag1 = 1

        #Milstein
        S2 = S2 + (r-q)*S1*dt + sig*S2*np.sqrt(dt)*z + 0.5*sig*sig*S2*dt*(z*z-1)

        if flag2 == 0:
            if S2 > Ub or S2 < Lb:
                flag2 = 1


    if flag1 == 0:
        runningSum1 += rebate

    if flag2 == 0:
        runningSum2 += rebate

    runningSum1b += max(S1-K,0)
    runningSum2b += max(S2-K,0)




premium1 = np.exp(-r*maturity)*runningSum1/N
premium2 = np.exp(-r*maturity)*runningSum2/N

premium1b = np.exp(-r*maturity)*runningSum1b/N
premium2b = np.exp(-r*maturity)*runningSum2b/N
```

```python
print(premium1)
print(premium2)

C = BMS_price('call', spot, K, r, q, sig, maturity)
print('C = ' + str(C))

print(premium1b)
print(premium2b)
```
