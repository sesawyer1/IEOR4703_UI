### Quasi-Monte Carlo

```python
import numpy as np
from time import time
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import qmc
```

```python
n = 2000
np.random.seed(645712)
```

### Regular Monte Carlo

```python
U0 = np.random.rand(n,2)
plt.figure()
plt.scatter(U0[:,0], U0[:,1], s=1, color='k')
plt.grid(True)
plt.title('Uniform random variables')
plt.show()
```

### Quasi-Monte Carlo

```python
halton_gen = qmc.Halton(d=2, scramble=False)
U1 = halton_gen.random(n=n)
plt.figure()
plt.scatter(U1[:, 0], U1[:, 1], s=1, c='k')
plt.grid(True)
plt.title('2-dim Halton sequence')
plt.show()
```

We want to estimate $\int_0^1 \int_0^1 x^3 (1+y^2) \, \mathrm{d}x \, \mathrm{d}y = \frac{1}{3}$.

```python
f0 = (U0[:,0]**3)*(1 + U0[:,1]**2)
I0 = np.mean(f0)
f1 = (U1[:,0]**3)*(1 + U1[:,1]**2)
I1 = np.mean(f1)

print('True value: 1/3')
print('MC : %f' % I0)
print('QMC: %f' % I1)
```
