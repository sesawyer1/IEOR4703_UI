### An example where Gibbs sampling fails

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 5000

s = np.random.rand()
if s < 0.5:
    x1 = -np.random.rand()
    x2 = -np.random.rand()
else:
    x1 = np.random.rand()
    x2 = np.random.rand()

x1Array = np.zeros(m)
x2Array = np.zeros(m)

# Burn-in
b = 5

for i in range(b + m):
    if i%2 == 0:
        if x2 < 0:
            x1 = -np.random.rand()
        else:
            x1 = np.random.rand()
    else:
        if x1 < 0:
            x2 = -np.random.rand()
        else:
            x2 = np.random.rand()

    if i >= b:
        x1Array[i-b] = x1
        x2Array[i-b] = x2
```

### Assessment by eye-balling

```python
plt.figure(figsize = (8, 6))
# domain
plt.plot([-1, -1], [0, -1], color='C0', linewidth=3)
plt.plot([1, 1], [0, 1], color='C0', linewidth=3)
plt.plot([-1, 1], [0, 0], color='C0', linewidth=3)
plt.plot([-1, 0], [-1, -1], color='C0', linewidth=3)
plt.plot([1, 0], [1, 1], color='C0', linewidth=3)
plt.plot([0, 0], [-1, 1], color='C0', linewidth=3)
# gibbs draws
plt.plot(x1Array, x2Array, color='C1', linewidth=1)
plt.show()
```
