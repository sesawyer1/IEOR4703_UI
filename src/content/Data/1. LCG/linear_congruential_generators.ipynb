### Linear Congruential Generator Example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 2**31 - 1
a = 7**5
c = 0

#seed
x = 98654

n = 100000
```

```python
rand_U = np.zeros(n)

for i in range(n):
    x = (a*x + c) % m
    rand_U[i] = 1.*x / (m + 1)
```

```python
plt.hist(rand_U, bins=10)
plt.show()
```
