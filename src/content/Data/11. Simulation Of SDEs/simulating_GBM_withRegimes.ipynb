### Simulation of Geometric Brownian Motion (GBM)


```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
spot = 500
maturity = 2
m = 500
l = 5
dt = maturity/m

# different vols
sig = [0.05, 0.15, 0.25, 0.35, 0.50]
mu = 0.20
```

```python
S = np.zeros((m + 1, l))
t = np.arange(0, maturity+dt, dt)
```

```python
np.random.seed(341166)
```

```python
z = np.random.randn(m)
S[0,:] = spot
```

```python
S[0,:] = spot
for j in range(l):
    for i in range(m):
        # exact solution
        S[i + 1,j] = S[i,j] * np.exp((mu - sig[j]*sig[j]/2)*dt + sig[j]*np.sqrt(dt)*z[i])
```

```python
fig = plt.figure(figsize=(18,6))
for j in range(l):
    plt.plot(t, S[:,j])
plt.show()
```

```python
z = np.random.randn(m)
S = np.zeros((m + 1, l))
S[0,:] = spot
```

```python
# regime switch
S[0,:] = spot
for j in range(l):
    for i in range(m):
        if i%50 == 0:
            k = random.randint(0, l-1)
            sig1 = sig[k]
        S[i + 1,j] = S[i,j] * np.exp((mu - sig1*sig1/2)*dt + sig1*np.sqrt(dt)*z[i])
```

```python
fig = plt.figure(figsize=(18,6))
for j in range(l):
    plt.plot(t, S[:,j])
plt.show()
```
