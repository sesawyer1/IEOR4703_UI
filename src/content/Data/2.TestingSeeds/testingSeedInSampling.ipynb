### Changing and testing random seeds

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the number of bins
num_bins = 10

# Compute the bin edges
bins = np.linspace(0, 1, num_bins + 1)
```

Testing seeds with Numpy function **np.random.seed**

```python
def plot_histogram(samples, num_bins=10):
    
    # Use np.histogram to bin the samples
    counts, bin_edges = np.histogram(samples, bins=bins)

    # Print the bin counts and bin edges
    print("Bin Counts: ", counts)
    print("Bin Edges: ", bin_edges)

    # Plotting the histogram
    plt.hist(samples, bins=bins, edgecolor='black', alpha=0.7)
    plt.title('Histogram of Binned Samples')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.xticks(bin_edges)  # Set the x-ticks to match the bin edges
    plt.grid(True)

    # Show the plot
    plt.show()
```

```python
np.random.seed(12345)
u = np.random.rand(100)
print(u[:5])
plot_histogram(u, num_bins=10)
```
