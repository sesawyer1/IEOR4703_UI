# IEOR 4703 -- Monte Carlo Simulation for FE


### Gibbs Sampling

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Binomial-Beta

```python
n = 40
alpha = 0.5
beta = 0.6
```

### Initial value

```python
x_0 = 0
theta_0 = np.random.rand()
```

### Gibbs sampling

```python
m = 1000

xArray = np.zeros(m+1)
thetaArray = np.zeros(m+1)

thetaArray[0] = theta_0
xArray[0] = x_0

freqs = np.zeros(n+1)

for i in range(1, m+1):
    theta = np.random.beta(alpha + xArray[i-1], beta + n - xArray[i-1])
    x = np.random.binomial(n, theta)

    xArray[i] = x
    thetaArray[i] = theta

    freqs[x] = freqs[x] + 1

freqs2 = freqs / np.sum(freqs)

# Some plots

plt.figure(figsize=[10, 10])

plt.subplot(3, 1, 1)
plt.plot(xArray)

plt.subplot(3, 1, 2)
plt.plot(thetaArray)

plt.subplot(3, 1, 3)
plt.bar(np.arange(n+1), freqs2)

plt.show()
```
