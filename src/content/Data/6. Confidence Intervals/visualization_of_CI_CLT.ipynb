## Visualization of confidence intervals via CLT

We wish to estimate the following integral via simulation:

$\int_0^1 \mathrm{e}^x \; \mathrm{d}x = \mathrm{e} - 1$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import scipy as sp
from time import time
```

Exact solution

```python
theta = np.exp(1) - 1
print('theta = %3.4f' % theta)
```

Approximate solution via simulation

```python
np.random.seed(11742567)
np.random.seed(21742567)
```

```python
# limits for plots
lowerB = theta - 0.5
upperB = theta + 0.5
boundY = 0.1

nN = [20, 50, 100, 200]
sizeN = len(nN)

alpha = np.array([0.01, 0.05, 0.5])
# compute quantiles
z = sp.stats.norm.ppf(1 - alpha / 2)
sizeM = len(alpha)

theta_hat_vec = np.zeros(sizeN)
lb_mat = np.zeros([sizeN, sizeM])
ub_mat = np.zeros([sizeN, sizeM])

for i in range(sizeN):
    
    N = nN[i]
    U = np.random.rand(N) # U~Uniform(0,1)
    Y = np.exp(U)
    theta_hat = np.mean(Y)
    sig2_N = np.var(Y, ddof=1)
    
    theta_hat_vec[i] = theta_hat
    lb_mat[i] = theta_hat - z * np.sqrt(sig2_N / N)
    ub_mat[i] = theta_hat + z * np.sqrt(sig2_N / N)    
```

Plot the confidence intervals

```python
fig, axs = plt.subplots(sizeN, sizeM, figsize=(15,12))
fig.subplots_adjust(hspace = .5, wspace = 1)

for i in range(sizeN):
    for j in range(sizeM):
        axs[i, j].plot([lb_mat[i,j], ub_mat[i,j]], [0,0], 'o', color='red')
        axs[i, j].plot(theta, 0, 'o', color='black')
        axs[i, j].set_title('N = %i, CI = %2.0f%s' % (nN[i], 100*(1-alpha[j]), '%'))
        axs[i, j].set_xlim([lowerB, upperB])
        axs[i, j].set_ylim([-boundY, boundY])
        
plt.show()
```
